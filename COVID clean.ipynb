```python
import pandas as pd
```

```python
cases = pd.read_csv('csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv')
cases.head()
```

```python
cases.tail()
```

```python
cases = cases.drop(['UID', 'iso2', 'iso3', 'code3'], axis=1)
cases = cases.dropna()
cases = cases.drop(cases.columns[3:60], axis=1)
cases.head()
```

```python
cases.tail()
```

```python
cases = cases.melt(id_vars=['FIPS', 'Admin2', 'Province_State'], value_vars=cases.columns[3:])
cases.columns = ['FIPS', 'County', 'State', 'Date', 'Cases']
cases['FIPS'] = cases['FIPS'].astype(int).astype(str)
cases['FIPS'] = cases['FIPS'].apply(lambda x: '0' + x if len(x) == 4 else x)
cases.head()
```

```python
cases.loc[0]
```

```python
cases.tail()
```

```python
deaths = pd.read_csv('csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv')
deaths.head()
```

```python
deaths.tail()
```

```python
deaths = deaths.drop(['UID', 'iso2', 'iso3', 'code3', 'Country_Region', 'Lat', 'Long_', 'Combined_Key'], axis=1)
deaths = deaths.dropna()
deaths = deaths.drop(deaths.columns[4:57], axis=1)
deaths.head()
```

```python
deaths.tail()
```

```python
deaths = deaths.melt(id_vars=['FIPS', 'Admin2', 'Province_State', 'Population'], value_vars=deaths.columns[4:])
deaths.columns = ['FIPS', 'County', 'State', 'Population', 'Date', 'Deaths']
deaths['FIPS'] = deaths['FIPS'].astype(int).astype(str)
deaths['FIPS'] = deaths['FIPS'].apply(lambda x: '0' + x if len(x) == 4 else x)
deaths.head()
```

```python
deaths.tail()
```

```python
cases.to_csv('csse_covid_19_data/csse_covid_19_time_series/covid19_cases_0809.csv')
deaths.to_csv('csse_covid_19_data/csse_covid_19_time_series/covid19_deaths_0809.csv')
```
